## **2. Xá»­ lÃ½, lÃ m sáº¡ch ná»™i dung comments**

```python
import re

# Sá»­ dá»¥ng danh sÃ¡ch Ä‘á»ƒ lÆ°u trá»¯ táº¥t cáº£ cÃ¡c giÃ¡ trá»‹ "message"
message_list = []

# Duyá»‡t qua danh sÃ¡ch JSON vÃ  thu tháº­p giÃ¡ trá»‹ "message"
for comment in data_comments:
    if 'message' in comment:
        message = comment['message']
        # Loáº¡i bá» cÃ¡c biá»ƒu tÆ°á»£ng hoáº·c kÃ½ hiá»‡u khÃ´ng pháº£i chá»¯ cÃ¡i, sá»‘ hoáº·c khoáº£ng tráº¯ng
        message = re.sub(r'[^\w\s]', '', message)
        # Loáº¡i bá» khoáº£ng tráº¯ng khÃ´ng cáº§n thiáº¿t á»Ÿ Ä‘áº§u vÃ  cuá»‘i cá»§a message
        message = message.strip()
        # ÄÃ³ng message trong dáº¥u ngoáº·c kÃ©p (")
        message = f'"{message}"'
        message_list.append(message)

# GhÃ©p cÃ¡c giÃ¡ trá»‹ "message" láº¡i thÃ nh má»™t chuá»—i, cÃ¡ch nhau bá»Ÿi dáº¥u pháº©y
combined_messages = ', '.join(message_list)

# In ra chuá»—i káº¿t quáº£
print(f"Tá»•ng sá»‘ bÃ¬nh luáº­n:", len(data_comments))
print(combined_messages)
```

# ***3.1. LÃ m sáº¡ch dá»¯ liá»‡u tuá»•i***



```python
import csv
from datetime import datetime

csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/all_data_file/user_data.csv'
output_csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/fresh_data_clean/user_age.csv'

def calculate_age(birthday):
    if not birthday or birthday.strip() == '':
        return 'khÃ´ng xÃ¡c Ä‘á»‹nh'
    else:
        try:
            date_obj = datetime.strptime(birthday, "%m/%d/%Y")
            current_year = datetime.now().year
            age = current_year - date_obj.year
            return age
        except ValueError:
            return 'khÃ´ng xÃ¡c Ä‘á»‹nh'

# Khá»Ÿi táº¡o danh sÃ¡ch trá»‘ng Ä‘á»ƒ lÆ°u trá»¯ tuá»•i
age_list = []

# Má»Ÿ tá»‡p CSV vÃ  Ä‘á»c ná»™i dung
with open(csv_file_path, 'r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        # Truy xuáº¥t thÃ´ng tin birthday tá»« dÃ²ng CSV vÃ  tÃ­nh tuá»•i
        birthday = row['birthday']
        age = calculate_age(birthday)
        age_list.append(age)

# Ghi danh sÃ¡ch tuá»•i vÃ o tá»‡p CSV má»›i
with open(output_csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['age'])  # Ghi tiÃªu Ä‘á» cho cá»™t tuá»•i
    for age in age_list:
        csv_writer.writerow([age])

print(f'Dá»¯ liá»‡u tuá»•i Ä‘Ã£ Ä‘Æ°á»£c ghi vÃ o tá»‡p {output_csv_file_path}')
```

# ***3.2. LÃ m sáº¡ch dá»¯ liá»‡u giá»›i tÃ­nh***

```python
import csv

csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/all_data_file/user_data.csv'
output_csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/fresh_data_clean/user_gender.csv'

# Khá»Ÿi táº¡o danh sÃ¡ch trá»‘ng Ä‘á»ƒ lÆ°u trá»¯ thÃ´ng tin giá»›i tÃ­nh
gender_list = []

# Má»Ÿ tá»‡p CSV vÃ  Ä‘á»c ná»™i dung
with open(csv_file_path, 'r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        # TrÃ­ch xuáº¥t giÃ¡ trá»‹ giá»›i tÃ­nh tá»« má»—i hÃ ng vÃ  kiá»ƒm tra
        gender = row['gender']
        if gender == 'male' or gender == 'female':
            gender_list.append(gender)
        else:
            gender_list.append('KhÃ´ng xÃ¡c Ä‘á»‹nh')

# Ghi danh sÃ¡ch giá»›i tÃ­nh vÃ o tá»‡p CSV má»›i
with open(output_csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['gender'])  # Ghi tiÃªu Ä‘á» cho cá»™t giá»›i tÃ­nh
    for gender in gender_list:
        csv_writer.writerow([gender])

print(f'Dá»¯ liá»‡u giá»›i tÃ­nh Ä‘Ã£ Ä‘Æ°á»£c ghi vÃ o tá»‡p {output_csv_file_path}')
```

## ***3.3. LÃ m sáº¡ch dá»¯ liá»‡u nÆ¡i sinh sá»‘ng***

```python
import csv
import json

csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/all_data_file/user_data.csv'
output_csv_file_path = '/content/drive/MyDrive/project_facebook_bigdata/fresh_data_clean/user_location.csv'

# Má»Ÿ tá»‡p user_data.csv
with open(csv_file_path, 'r') as file:
    # Äá»c ná»™i dung cá»§a tá»‡p CSV
    csv_reader = csv.DictReader(file)

    # Khá»Ÿi táº¡o danh sÃ¡ch Ä‘á»ƒ lÆ°u cÃ¡c giÃ¡ trá»‹ 'name' tá»« cá»™t 'location'
    names_in_location = []

    # Láº·p qua tá»«ng dÃ²ng trong tá»‡p CSV
    for row in csv_reader:
        # Kiá»ƒm tra xem chuá»—i 'location' cÃ³ Ä‘Ãºng Ä‘á»‹nh dáº¡ng JSON hay khÃ´ng
        try:
            location_data = json.loads(row['location'].replace("'", "\""))
            if 'name' in location_data:
                names_in_location.append(location_data['name'])
            else:
                names_in_location.append("khÃ´ng xÃ¡c Ä‘á»‹nh")
        except json.JSONDecodeError:
            names_in_location.append("khÃ´ng xÃ¡c Ä‘á»‹nh")

# Ghi danh sÃ¡ch giÃ¡ trá»‹ 'name' vÃ o tá»‡p CSV má»›i
with open(output_csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['TÃªn'])
    for name in names_in_location:
        csv_writer.writerow([name])

print(f'Dá»¯ liá»‡u Ä‘Ã£ Ä‘Æ°á»£c ghi vÃ o tá»‡p ',output_csv_file_path)
```
