```python

# Đọc tệp JSON
with open(file_json_data, 'r') as file:
    data = json.load(file)

# Tạo danh sách rỗng để lưu trữ thông tin name và id
names_and_ids = []

# Lặp qua các đối tượng trong tệp JSON và trích xuất name và id
for item in data:
    if 'from' in item and 'name' in item['from'] and 'id' in item['from']:
        name = item['from']['name']
        user_id = item['from']['id']
        names_and_ids.append({'name': name, 'id': user_id})

print(len(names_and_ids))
names_and_ids
```
